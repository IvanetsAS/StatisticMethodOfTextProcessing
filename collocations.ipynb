### Task
Взять **nltk** и натравить его на большой корпус текстов и вывести первые коллокации по нескольким метрикам


```python
import os

path = "Corpus/School"
file_list = os.listdir(path)

unprocessed_corpus = ""

for file in file_list:
    f = open(path + "/" + file, "r")
    unprocessed_corpus += f.read()
```

```python
from nltk.tokenize import word_tokenize

tokenized_corpus = word_tokenize(unprocessed_corpus)
```

```python
def preprocess_word(word):
    word = word.lower()
    return morph.parse(word)[0].normal_form
```

```python
from nltk.corpus import stopwords
import string

stopWords = set(stopwords.words('russian'))
stopWords = stopWords | set(string.punctuation)
```

```python
import pymorphy2

morph = pymorphy2.MorphAnalyzer()

normalized_corpus = []
i = 0
length = len(tokenized_corpus)
for word in tokenized_corpus:
    if word not in stopWords:
        normalized_corpus.append(preprocess_word(word))
    
    i += 1
    percent = (i / length) * 100
    if i % 100000 == 0:
        print(i, " ", percent, "%")
```

```python
len(normalized_corpus)
```

```python
import nltk
from nltk.collocations import *

bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()

finder = BigramCollocationFinder.from_words(normalized_corpus)
```

```python
finder.apply_freq_filter(10)
```

```python
n_best_pmi = finder.nbest(bigram_measures.pmi, 100)
```

```python
n_best_pmi
```

```python
n_best_chi_sq = finder.nbest(bigram_measures.chi_sq, 100)
```

```python
n_best_dice = finder.nbest(bigram_measures.dice, 100)
```
